```python
import pyautogui
import time
import pyperclip

# entrar no navegador e baixar arquivo

pyautogui.PAUSE = 1

pyautogui.press('win')
pyautogui.write('chrome')
pyautogui.press('enter')
time.sleep(3)

link = "https://drive.google.com/drive/u/0/folders/1O8shYlSTJVGxjoxY4piFIDd6YDECkkjS"
pyperclip.copy(link)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')

time.sleep(7)

pyautogui.click(437, 341, clicks=2)

pyautogui.click(470, 413) # clica no arquivo

pyautogui.click(1658, 196) # clica nos 3 pontinhos

pyautogui.click(1428, 665) # clica para fazer download

time.sleep(10)

pyautogui.click(1890, 12)
```

```python
import pandas as pd
from IPython.display import display # importante

# ler a tabela

tabela = pd.read_excel("C:\\Users\\bitte\\Downloads\\Vendas - Dez.xlsx")

faturamento = tabela["Valor Final"].sum() # soma da coluna Valor Final
quantidade = tabela["Quantidade"].sum() # soma da coluna Quantidade

display(tabela)
```

```python
# enviar email

pyautogui.PAUSE = 1

pyautogui.press('win')
pyautogui.write('chrome')
pyautogui.press('enter')
pyautogui.click(1870, 53)
time.sleep(3)

link = "https://mail.google.com/mail/u/0/#inbox"
pyperclip.copy(link)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')


time.sleep(7)

pyautogui.click(43, 218) # clica no botão escreverchrome

pyautogui.write("rlaurentino96@gmail.com")
pyautogui.press('tab')
pyautogui.press('tab')

assunto = "Relatório de Vendas"
pyperclip.copy(assunto)
pyautogui.hotkey("ctrl", 'v')
pyautogui.press('tab')


texto_email = f"""
Prezados, bom dia

O faturamento de ontem foi de: R$ {faturamento:,.2f}
A quantidade de produtos foi de: {quantidade:,}

Abs
Rossana Silva
"""
pyperclip.copy(texto_email)
pyautogui.hotkey("ctrl", 'v')

pyautogui.hotkey('ctrl', 'enter')

pyautogui.alert("Fim da Automação. Seu computador já voltou a ser seu")
```
